```python
import gzip
import pandas as pd 
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
import re
import os
    
patternHTML = r"(<\/?)([a-zA-Z0-9])*(( [a-zA-Z]+\=\"[ \S]+\")+)?( )?(\/)?(>)"
regexpHTML = re.compile(patternHTML)

patternPre = r"(<pre)(( [a-zA-Z]+\=\"[ \S]+\")+)?(>)([\s\S])*(<\/pre>)"
regexpPre = re.compile(patternPre)

patternImg = r"(<img)(( [a-zA-Z]+\=\"[ \S]+\")+)(\/)?(>)"
regexpImg = re.compile(patternImg)

patternTable = r"(<table>)([\s\S])*(<\/table>)"
regexpTable = re.compile(patternTable)

patternUl = r"(<ul>)([\s\S])*(<\/ul>)"
regexpUl = re.compile(patternUl)

patternOl = r"(<ol>)([\s\S])*(<\/ol>)"
regexpOl = re.compile(patternOl)

patternScript = r"(<script>)([\s\S])*(<\/script>)"
regexpScript = re.compile(patternScript)

patternStyle = r"(<style>)([\s\S])*(<\/style>)"
regexpStyle = re.compile(patternStyle)

patternName = r"@( )?\S+"
regexpName = re.compile(patternName)

patternUrl = r"http\S+"
regexpUrl = re.compile(patternUrl)

patternLateX = r"(\$+)(?:(?!\1)[\s\S])*\1"
regexpLateX = re.compile(patternLateX)

patternLateXbis = r"\\begin\{(?:(?!\})[\S])+\}([\s\S])*(\\end\{(?:(?!\})[\S])+\})"
regexpLateXbis = re.compile(patternLateXbis)

patternPunctuation = r"([\;\,\"\'\(\)\&\$\+\-\_\%\:\\\*\@\^\[\]\/\=\`\<\>])"
regexpPunctuation = re.compile(patternPunctuation)

patternDigit = r"(\b\d+(?:\.\d+)?)"
regexpDigit = re.compile(patternDigit)

patternMix = r"\w*[0-9]\w*"
regexpMix = re.compile(patternMix)

patternVoid = r"\s+"
regexpVoid = re.compile(patternVoid)

lst = []

with gzip.open('stackexchange_812k.csv.gz', 'r') as f:
    data = pd.read_csv(f)


def addWhiteSpace(match_obj):
    return(" "+match_obj.group(1)+" ")

def emptySpaces(match_obj):
    return(" ")

def append_line(file_name, text_to_append):
    with open(file_name, "a+") as file:
        if os.stat(file_name).st_size != 0:
            file.write("\n")
        file.write(text_to_append)
        
for line in data.itertuples():
    lineClean=line.text
    if regexpPre.search(lineClean):
        lineClean = regexpPre.sub(emptySpaces,lineClean)
    if regexpImg.search(lineClean):
        lineClean = regexpImg.sub(emptySpaces,lineClean)
    if regexpTable.search(lineClean):
        lineClean = regexpTable.sub(emptySpaces,lineClean)
    if regexpUl.search(lineClean):
        lineClean = regexpUl.sub(emptySpaces,lineClean)
    if regexpOl.search(lineClean):
        lineClean = regexpOl.sub(emptySpaces,lineClean)
    if regexpScript.search(lineClean):
        lineClean = regexpScript.sub(emptySpaces,lineClean)
    if regexpStyle.search(lineClean):
        lineClean = regexpStyle.sub(emptySpaces,lineClean)
    if regexpHTML.search(lineClean):
        lineClean = regexpHTML.sub(emptySpaces,lineClean)
    if regexpUrl.search(lineClean):
        lineClean = regexpUrl.sub(emptySpaces,lineClean)
    if regexpLateX.search(lineClean):
        lineClean = regexpLateX.sub(emptySpaces,lineClean) 
    if regexpLateXbis.search(lineClean):
        lineClean = regexpLateXbis.sub(emptySpaces,lineClean)  
    if regexpName.search(lineClean):
        lineClean = regexpName.sub(emptySpaces,lineClean)    
    if regexpDigit.search(lineClean):
        lineClean = regexpDigit.sub(emptySpaces,lineClean) 
    if regexpMix.search(lineClean):
        lineClean = regexpMix.sub(emptySpaces,lineClean) 
    lineClean = re.compile(r"(\W)").sub(addWhiteSpace,lineClean)
    if regexpPunctuation.search(lineClean):
        lineClean = regexpPunctuation.sub(emptySpaces,lineClean)
    if regexpVoid.search(lineClean):
        lineClean = regexpVoid.sub(emptySpaces,lineClean) 
    lineClean = lineClean.lower()
    lst.append(lineClean)
                
data_cleaned = data.copy()
data_cleaned["cleaned_text"]=lst
data_cleaned.to_csv("submitDataset.csv")
```
